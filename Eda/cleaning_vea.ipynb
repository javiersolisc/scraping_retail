<h2>$1: Eliminación \ de \ columnas \ que \ no \ usaremos$</h2>

```python
data2 = data.drop(columns = ["alto", "alto con base", "ancho", "conexiÃ³n bluetooth", "enfoque automatico", "incluye", "procesador", "profundidad", "radio", "resolucion video", "vendido por", "memoria expandible", "conexion wi fi"])
```

<h2>$1.1: Cambio\ de \ nombre\ al\ encabezado\ de\ las\ columnas$</h2>

```python
data3=data2.rename({'bateria':'bat', 'marca0':'brand', 'camara frontal':'cam_front', 'camara principal':'cam_main', 'color':'color', 'tecnologia celular':'conect', 'nucleos del procesador':'cores', 'velocidad del procesador':'cpu_vel', 'memoria interna':'mem_int', 'memoria ram':'mem_ram', 'modelo':'phone', 'sistema operativo':'os', 'nombre del producto':'product', 'resolucion de pantalla':'screen_res', 'tamaÃ±o pantalla celulares':'screen_size', 'tipo de pantalla':'screen_type', 'dual sim':'sim', 'garantia':'warranty', 'resistente al agua':'water_res', 'precio normal':'price_norm', 'precio oferta':'price_off'},axis=1)
```

<h2>$1.2: Crear\  nueva\  columna\  con\  valores\  de\  la\  data$</h2>

```python
data3 = data3.copy()
data3['condition'] = data3['product'].str.extract('(Reacondicionado|REACONDICIONADO)')
```

```python
#En la columna Condition remplazar los valores nulos por "nuevo"
data3['condition'].fillna('nuevo', inplace=True)
```

```python
data3
```

<h2>$1.3: Ordenar\ columnas$</h2>

```python
data4=data3[['bat','brand','cam_front','cam_main','color','condition','conect','cores','cpu_vel','mem_int','mem_ram','phone','os','product','screen_res','screen_size','screen_type','sim','warranty','water_res','price_norm','price_off']]
```

```python
data4
```

<h2>$2: Remplazar\ valores$</h2> 

```python
data4 = data4.copy()
data4.loc[:, 'bat'] = data4['bat'].str.replace('([a-zA-Z]+)', '', regex=True)
```

```python
data4.loc['bat'] = data4['bat'].str.replace('(Ã¡|Ã\xad|-|,|á|í|Ã¡)', '', regex=True) 
```

```python
data4['cam_front'] = data4['cam_front'].str.replace('([a-zA-Z]+)', '', regex=True)
```

```python
 data4['cam_front'] = data4['cam_front'].str.replace('(Ã¡|Ã\xad|-|,|/|Ã³| Æ’|[()])', '', regex=True)   
```

```python
 data4['cam_main'] = data4['cam_main'].str.replace('([a-zA-Z]+)', '', regex=True)
```

```python
data4['cam_main'] = data4['cam_main'].str.replace('(Ã¡|Ã\xad|-|/|Ã³| Æ’|[()]|Âµ|Â°|[|])', '', regex=True)   
```

```python
data4['cam_main'] = data4['cam_main'].str.replace('([+])', ' ', regex=True) 
```

```python
data4['cam_main'] = data4['cam_main'].str.split(" ", expand=True, n=1)[0]
```

```python
data4['cpu_vel'] = data4['cpu_vel'].str.replace('([a-zA-Z]+)', '', regex=True)
```

```python
data4['cpu_vel'] = data4['cpu_vel'].str.replace('(@|Ã—|\tÃ\xad-|&|[()])', ' ', regex=True)  
```

```python
data4['mem_int'] = data4['mem_int'].str.replace('([a-zA-Z]+)', '', regex=True)
```

```python
data4['mem_ram'] = data4['mem_ram'].str.replace('([a-zA-Z]+)', '', regex=True)
```

```python
data4['screen_res'] = data4['screen_res'].str.replace('([a-zA-Z]+)', ' ', regex=True)
```

```python
data4['screen_res'] = data4['screen_res'].str.replace('(Ã\xad|Ã³|Ã—|Â·)', ' ', regex=True)
```

```python
data4['screen_size'] = data4['screen_size'].str.replace('([a-zA-Z]+)', '', regex=True)
```

```python
 data4['screen_size'] = data4['screen_size'].str.replace('(["])', '', regex=True)
```

```python
data4['warranty'] = data4['warranty'].str.replace('([a-zA-Z]+)', '', regex=True)
```

```python
data4['warranty'] = data4['warranty'].str.replace('(Ã±|Ã‘|Ã¡|Ã||x|ñ|Ñ)', '', regex=True)
```

```python
 data4['warranty'] = data4['warranty'].str.replace('(0)', '', regex=True)  
```

```python
#Cambiar 1 año a 12 meses
data4['warranty'] = data4['warranty'].str.replace('(1 )', '12', regex=True)
```

```python
data4['price_norm'] = data4['price_norm'].str.replace('([a-zA-Z]+)', '', regex=True)
```

```python
data4['price_off'] = data4['price_off'].str.replace('([a-zA-Z]+)', '', regex=True)
```

```python
data4['price_norm'] = data4['price_norm'].str.replace('(/|-)', '', regex=True)
```

```python
data4['price_off'] = data4['price_off'].str.replace('(/|-)', '', regex=True)
```

```python
data4['price_norm'] = data4['price_norm'].str.replace('([,])', '', regex=True)
```

```python
data4['price_off'] = data4['price_off'].str.replace('([,])', '', regex=True)
```

<h2>$3: Eliminar\ valores$</h2> 

```python
#Eliminar los ultimos 3 valores "46%"
data4['price_norm'] = data4['price_norm'].str[:-3]
```

```python
data4['price_off'] = data4['price_off'].str[:-5]
```

```python
data4['product'] = data4['product'].str.replace('(Reacondicionado|REACONDICIONADO|Smartphone|Celular|Teléfono|Apple)','', regex=True)
```

```python
data4['product'] = data4['product'].apply(lambda x: x[:x.find('GB')]if isinstance(x, str) else float('nan'))
```

```python
data4['product'] = data4['product'].apply(lambda x: x[:x.find('"')] if isinstance(x, str) else float('nan'))
```

```python
data4['product'] = data4['product'].apply(lambda x: x[:x.find('.')]if isinstance(x, str) else float('nan'))   
```
