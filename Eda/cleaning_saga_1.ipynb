<h2>1: Eliminación  de  columnas  que  no  usaremos</h2>

```python
data = data.drop(['alto', 
                    'ancho',
                    'conexion bluetooth',
                    'gps integrado',
                    'lector de huella',
                    'marca',
                    'memoria expandible',
                    'peso',
                    'profundidad',
                    'tipo'], axis=1)
```

```python
data
```

<h2>1.1: Cambio de  nombre al encabezado de las columnas</h2>

```python
# Renombramos las columnas segun nuestro diccionario
data = data.rename({'bateria':'bat',
                    'camara frontal':'cam_front',
                    'camara posterior':'cam_main',
                    'condicion del producto':'condition',
                    'conectividad':'conect',
                    'garantia del provedor':'warranty',
                    'marca0':'brand',
                    'memoria interna':'mem_int',
                    'memoria ram':'mem_ram',
                    'modelo':'model',
                    'nombre del producto':'product',
                    'MODELO DEL PRODUCTO':'modelo',
                    'nucleos del procesador':'cores',
                    'precio normal':'price_norm',
                    'precio oferta':'price_off',
                    'resistente al agua':'water_res',
                    'sistema operativo':'os',
                    'tamaño de la pantalla':'screen_size',
                    'tipo de pantalla':'screen_type',
                    'velocidad del procesador':'cpu_vel'}, axis=1)
```

<h2>1.2: Crear  nuevas  columnas  con  valores  de  la  data</h2>

```python
data['warranty'] = None
```

```python
data['condition_new'] = data['product'].str.extract('(Reacondicionado|REACONDICIONADO)')
data['condition'].fillna(value=data['condition_new'], inplace=True)
data.drop(['condition_new'], axis=1, inplace=True)
```

```python
data = data.copy()
data['color'] = data['product'].str.extract('(white|blanco|black|negro|)')
```

```python
data = data.copy()
data['sim'] = data['product'].str.extract('(doble|double|single|)')
```

```python
data
```

<h2>1.3: Ordenar columnas</h2>

```python
data=data[['bat','brand','cam_front','cam_main','color','condition','conect','cores','cpu_vel','mem_int','mem_ram','model','os','product','screen_size','screen_type','sim','warranty','water_res','price_norm','price_off']]
```

```python
data
```

<h2>2: Remplazar valores</h2> 

```python
data = data.copy()
data.loc[:, 'bat'] = data['bat'].str.replace('([a-zA-Z]+)', '', regex=True)

data['bat'] = data['bat'].str.replace('(Ã¡|Ã\xad|-|,|)', '', regex=True)
```

```python
data['cam_front'] = data['cam_front'].str.replace('([a-zA-Z]+)', '', regex=True)
data['cam_front'] = data['cam_front'].str.replace('(Ã¡|Ã\xad|-|,|/|Ã³| Æ’|[()])', '', regex=True)   
```

```python
data['cam_main'] = data['cam_main'].str.replace('([a-zA-Z]+)', '', regex=True)
data['cam_main'] = data['cam_main'].str.replace('(Ã¡|Ã\xad|-|/|Ã³| Æ’|[()]|Âµ|Â°|[|])', '', regex=True)
data['cam_main'] = data['cam_main'].str.replace('([+])', ' ', regex=True)
data['cam_main'] = data['cam_main'].str.split(" ", expand=True, n=1)[0]
```

```python
data['cpu_vel'] = data['cpu_vel'].str.replace('([a-zA-Z]+)', '', regex=True)
```

```python
data['mem_int'] = data['mem_int'].str.replace('([a-zA-Z]+)', '', regex=True)
```

```python
data['mem_ram'] = data['mem_ram'].str.replace('([a-zA-Z]+)', '', regex=True)
```

```python
data['screen_size'] = data['screen_size'].str.replace('([a-zA-Z]+)', '', regex=True)
data['screen_size'] = data['screen_size'].str.replace('(["])', '', regex=True)
```

```python
data['warranty'] = data['warranty'].str.replace('([a-zA-Z]+)', '', regex=True)
data['warranty'] = data['warranty'].str.replace('(Ã±|Ã‘|Ã¡|Ã||x)', '', regex=True)
data['warranty'] = data['warranty'].str.replace('(0)', '1', regex=True)  
data['warranty'] = data['warranty'].str.replace('(12|13)', '1', regex=True)
```

```python
data['price_norm'] = data['price_norm'].str.replace('([a-zA-Z]+)', '', regex=True)
data['price_norm'] = data['price_norm'].str.replace('(Ã±|Ã‘|Ã¡|Ã||x|,|/)', '', regex=True)
```

```python
data['price_off'] = data['price_off'].str.replace('([a-zA-Z]+)', '', regex=True)
data['price_off'] = data['price_off'].str.replace('(Ã±|Ã‘|Ã¡|Ã||x|,|/)', '', regex=True)
data['price_off'] = data['price_off'].str.split(" ", expand=True, n=1)[0]
```

```python
data
```

<h2>3: Eliminar valores</h2> 

```python
data['product'] = data['product'].apply(lambda x: x[:x.find('GB')]if isinstance(x, str) else float('nan'))
```

```python
data
```

<h2>3: Convertir objects a floats</h2>

```python
data['price_norm'] = data['price_norm'].replace(' ', np.nan)
```

```python
data['price_norm'] = pd.to_numeric(data['price_norm'])
```

```python
data['price_off'] = pd.to_numeric(data['price_off'])
```
